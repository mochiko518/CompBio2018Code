```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# データの読み込み
data = np.genfromtxt("./CB.csv", dtype="int",delimiter=",")
```

```python
# プロット
plt.figure( figsize=(5, 5), dpi=120)
plt.matshow(data, cmap=plt.cm.gray, fignum=1)
plt.show()
```
